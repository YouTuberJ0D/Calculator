```python
#Calculator

def add(x,y):
  return x+y
def sub(x,y):
  return x-y
def mul(x,y):
  return x*y
def div(x,y):
  return x/y
def pow(x,y):
  return x**y
x  = int(input("Tell the number "))
y  = int(input("Tell the number "))
operation =  input("Tell the operation ")
if operation == "+" :
  print( add(x,y))
elif operation == "-":
  print(sub(x,y))
elif operation == "*":
  print(mul(x,y))
elif operation == "/":
  print(div(x,y))
elif operation == "^":
  print(pow(x,y))
```

```python
#oops
class shoes:
  def __init__(self,material,size):
    self.material = material
    self.size = size
nike = shoes("leather",6)
comet = shoes("cotton",7)

print(nike.size)
print(comet.material)
```
